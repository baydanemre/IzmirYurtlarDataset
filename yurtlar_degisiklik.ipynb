```python
import pandas as pd

yurtlar_excel = pd.read_excel('yurtlar_data_new_3.xlsx')

yurtlar_df_copy = yurtlar_excel.copy()
```

```python
df = pd.DataFrame(yurtlar_df_copy)
df.head()
```

```python
yurt_enlem = list()
yurt_boylam = list()
```

```python
for i in range(0,len(df)):
    enlem= df["yurt_koordinat"][i].split(",")[0]
    boylam = df["yurt_koordinat"][i].split(",")[1]
    yurt_enlem.append(enlem)
    yurt_boylam.append(boylam)
```

```python
df["yurt_enlem"] = yurt_enlem
df["yurt_boylam"] = yurt_boylam
```

```python
df.head()
```

```python
df.drop("yurt_koordinat", axis=1,inplace=True)
```

```python
df.head()
```

```python
dosya_adi = "izmir_yurtlar_DF.xlsx"
df.to_excel(dosya_adi,sheet_name="Yurt_Verileri", index=False)
```
